```python
import pandas as pd
from fbprophet import Prophet
# from vega_datasets import data
# import plotly.express as px
import fbprophet.plot as fp
# import plotly
```

```python
df = pd.read_csv('C:/Users/owner/Desktop/python/prophet/EURUSD1h.csv')
df.head()
```

```python
df['ds']=df['Date'].astype(str).str[:4]+'-'+df['Date'].astype(str).str[4:6]+'-'+df['Date'].astype(str).str[6:8]+' '+df['Timestamp']
df['y']=df['Close']
df = df.loc[:,['y','ds']]
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=300, freq='H')
```

```python
fcst = m.predict(future)
```

```python
fig = m.plot(fcst)
```
